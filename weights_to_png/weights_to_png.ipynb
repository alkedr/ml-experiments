```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "Qwen/Qwen2.5-0.5B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Get model weights as a tensor
model_weights = []
for param in model.parameters():
    model_weights.append(param.data.flatten())
weights_tensor = torch.cat(model_weights)
print(f"Total number of parameters: {weights_tensor.shape[0]:,}")

```

```python
# params = list(model.parameters())
# params.sort(key=lambda x: x.data.numel(), reverse=True)

# for param in params:
#     print(param.data.numel(), param.data.shape)


embeddings = model.model.embed_tokens.weight

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Convert embeddings to numpy array and normalize to 0-255 range for int8
embeddings_np = embeddings.detach().numpy()
embeddings_norm = ((embeddings_np - embeddings_np.min()) / (embeddings_np.max() - embeddings_np.min()) * 255).astype(np.uint8)


embeddings_norm = embeddings_norm[:200, :200]
embeddings_norm = np.repeat(embeddings_norm, 8, axis=0)
embeddings_norm = np.repeat(embeddings_norm, 8, axis=1)


img = Image.fromarray(embeddings_norm, mode="L")
img.save('my.png')
# img.show()

# np.sum(embeddings_norm, axis=1).shape

embeddings_norm[0]
```
